Content Based Recommendation system

```python
#Based on movie's plot summaries given in the overview. If our user gives us a movie title, we will suggest movies
#that shares similar plot

from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(min_df =3,#ignore terms that have document frequency less than this value(can be a typo)
                    max_features = None, #if not None,build a vocabulary that consider only top max features ordered by term frequency
                    strip_accents ='unicode',#removes accents and perform other character normalization during pre-processing
                    stop_words ='english',
                    analyzer ='word',
                    ngram_range =(1,3),
                    token_pattern = r' w{1,}')#only used when the analyzer is the word
```

```python
tfv_matrix = tf.fit_transform(df['overview'].values.astype('U'))

print(tfv_matrix)
```

```python
print(tfv_matrix.shape)
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tfv_matrix,tfv_matrix)
```

```python
print(sig[0])
```

```python
#Reverse mapping of indices and movie titles
indices = pd.Series(df.index,index = df['original_title']).drop_duplicates()

print(indices)
```

```python
print(indices['El Mariachi'])
```

```python
list(enumerate(sig[indices['El Mariachi']]))
```

```python
sorted(list(enumerate(sig[indices['El Mariachi']])),key = lambda x:x[1],reverse=True)
```

```python
def give_recommendations(title,sig= sig):
    idx = indices[title]

    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores,key=lambda x: x[1],reverse=True)

    #score of 10 most similar movies
    sig_scores = sig_scores[1:11]

    movie_indices = [i[0] for i in sig_scores]

    return df['original_title'].iloc[movie_indices]
```

```python
print(give_recommendations('Avatar'))
```

```python
print(give_recommendations('El Mariachi'))
```
